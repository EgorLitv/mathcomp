```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc

def function(x):
    result=-x**2-x+np.cos(x)+x*np.cos(x)
    return result

x=np.linspace(0,1,100)

plt.plot(x,function(x))
plt.show
```

```python
from scipy import optimize

optimize.newton(function,0.5)
```

```python
def function1(x):
    result=np.sin(x)/(1+x**2)
    return result
x=np.linspace(0,10,50)
zh=function1(x)
plt.plot(x,function1(x))
plt.show


plt.plot(x,np.gradient(zh,edge_order=1),'r*',x,np.gradient(zh,edge_order=2),'g-',x,np.gradient((np.gradient(zh,edge_order=2)),edge_order=1),'b--')
plt.show
```
